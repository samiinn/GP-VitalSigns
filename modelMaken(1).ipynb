```python
def dataMaker(version):
    df = pd.read_csv ('realData/uq_vsd_case' + version + '_trenddata.csv', error_bad_lines=False, index_col = False )

    df = df.loc[:580]
    temp = []

    time = df['RelativeTimeMilliseconds']
    pulse = df['Pulse']
    RR = df['RR']

    temp.append(time)
    temp.append(pulse)
    temp.append(RR)

    return temp
```

```python
data = []

for i in range(1,33):
    temp = str(i)
    if i < 10:
        data.append(dataMaker('0' + temp))
    else:
        data.append(dataMaker(temp))

#data now contains the first 580 values for each subject with, time, pulse and rr
```

```python
#here: loop through data list, fit a model for each attibute against time, and save those model (2 in total for each subject) in a separate list. 

#the final list should be a list of list, where the outer list is 32 long and each index contains two models for each subject.
```

```python
data[0][2]
```

```python
list_models = []
kernel = GPy.kern.RBF(input_dim=1)
for i in data:
    list_triple = []
    for j in i:
        data_arr = StandardScaler().fit_transform(j.values.reshape(-1,1))
        list_triple.append(np.array(data_arr))
        

    model_pulse = GPy.models.GPRegression(list_triple[0], list_triple[1], kernel)
    model_pulse.optimize(messages=True)
    model_pulse.optimize_restarts(num_restarts = 2)


    model_RR = GPy.models.GPRegression(list_triple[0], list_triple[2], kernel)
    model_RR.optimize(messages=True)
    model_RR.optimize_restarts(num_restarts = 2)
    
    time_pulse_model = model_pulse.to_dict()
    time_RR_model = model_RR.to_dict()
    list_models.append([time_pulse_model, time_RR_model])
        
```
